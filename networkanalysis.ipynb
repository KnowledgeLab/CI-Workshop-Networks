```python
%matplotlib inline
import pylab as plt
from collections import Counter
from igraph import * 
import numpy as np
import pandas as pd
import random

from IPython.display import Image

# a function for plotting graphs, in case igraph's plot does not work
def drawgraph(g,layout,vertex_label='',vertex_size=150,vertex_color='skyblue'):
    plt.figure(figsize=(10,10))
    X=np.array(layout.coords)
    for e in g.es:
        plt.plot(X[[e.source,e.target],0],X[[e.source,e.target],1],'grey',lw=1)
    plt.scatter(X[:,0],X[:,1],s=vertex_size,c=vertex_color)
    if vertex_label:
        for i in xrange(X.shape[0]):
            plt.text(X[i,0],X[i,1],vertex_label[i])
    plt.axis('off')
```

### 4. Community Detection

A community is widely regarded as a group of nodes that are densely connected with each other but loosely connected to nodes outside of the group. There is no formal definition of communities nor community detection. With different objectives there is a wide array of methods to find meanful partitions of nodes.

#### 4.1 Modularity

Given a partion of the network as communities, the modularity $Q$ of this partition is defined as:

$$ Q=\frac{1}{2m}\sum_{ij}(A_{ij}-\frac{k_ik_j}{2m})\delta(c_i,c_j), $$

where $c_i$ is the community node $i$ is in, and $\delta(c_i,c_j)=1$ if $c_i=c_j$ and $\delta(c_i,c_j)=0$ otherwise. Coarsely speaking, $Q$ counts the number of links in all communities compared to the number of links in a random network. 
  
Therefore, the goal of community detection is to find the partition that maximize the modularity $Q$.

```python
g=Graph.Read_Edgelist('Caltech.txt',directed=False)  # Load the facebook network of Caltech in September 2005. Traud et al.,Comparing Community Structure to Characteristics in Online Collegiate Social Networks,SIAM Review 2011 
g.delete_vertices(0)
g.simplify();
```

```python
with open('Caltech_info.txt') as infile: # Load information about which dorms people live in
    for i,line in enumerate(infile):
        g.vs[i]['dorm']=line.strip()
```

```python
comm=g.community_leading_eigenvector()
```

```python
palette=dict(zip(list(set(g.vs['dorm'])),
                 ['white','yellowgreen', 'gold', 'lightskyblue', 'lightcoral','cornsilk','silver','tomato','royalblue'])) 
plt.figure(figsize=(20,20))
for c in xrange(len(comm)):
    dorms=Counter(comm.subgraph(c).vs['dorm'])
    slices=dorms.values()
    colors=[palette[i] for i in dorms.keys()]
    explode=[0]*len(dorms)
    for i,k in enumerate(dorms.keys()):
        if k==dorms.most_common(1)[0][0]:
            explode[i]=0.1
    plt.subplot(len(comm)/3+1,3,c+1)
    plt.pie(slices, explode=explode, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
    plt.title('Community '+str(c))
    plt.axis('equal');
```

### 5. Link Prediction

Given an observed network, link prediction aim to solve two practical problems: find missing links or predict potential links.

#### 5.1 Feature-based methods

* Common neighbors: number of common neighbors between two nodes
* Jaccard similarity: common neighbors normalized by the size of the joint neighborhood
$$ Jaccard(i,j)=\frac{|\Gamma(i)\cap\Gamma(j)|}{|\Gamma(i)\cup\Gamma(j)|} $$
* Adamic/Adar: number of weighted common neighbors; each neighbors is discounted by its degree.
$$ AdamicAdar(i,j)=\sum_{z\in\Gamma(i)\cap\Gamma(j)}\frac{1}{\log|\Gamma(z)|} $$.
* Distance: length of shortest paths.
* Kats similarity: number of paths between two nodes but the contribution from long paths is discounted.
$$ Katz(i,j)=\sum_{l=1}^{\infty}\beta^l~|paths_{i,j}^l|~~~~or~~~~Katz=(I-\beta A)^{-1}-I$$.

* Rooted Pagerank: the similarty between nodes $i$ and $j$ is the stationary probability of visiting $j$ of a random walk that starts from $i$ and returns to $i$ with probability $1-\beta$ in every step.
$$ Pagerank=(1-\beta)(I-\beta D^{-1}A)^{-1} $$

```python
common_neighbors=np.zeros((g.vcount(),g.vcount()))
jaccard=np.zeros((g.vcount(),g.vcount()))
adamicadar=np.zeros((g.vcount(),g.vcount()))
for i in xrange(g.vcount()):
    for j in xrange(i+1,g.vcount()):
        common_neighbors[i,j]=len(np.intersect1d(g.neighbors(i),g.neighbors(j)))
        jaccard[i,j]=len(np.intersect1d(g.neighbors(i),g.neighbors(j)))*1.0/len(np.union1d(g.neighbors(i),g.neighbors(j)))
        adamicadar[i,j]=sum(1/np.log(np.intersect1d(g.neighbors(1),g.neighbors(2))))

distance=g.shortest_paths()
```

#### 5.2 Matrix Factorization

Link prediction is the engine of recommendar systems. There are normally two types of nodes (e.g., users and items) and links only exist between users and items representing, for example, which movie a user watched, what music a user like, what book a user read, and so on. The idea to predict a link between a user and an item she is not connected to is to consider what users similar to her like and what items she liked. This general framework is called collaborative filtering.

```python
Image('collaborativefiltering.png')  # Image from wikipedia:http://en.wikipedia.org/wiki/Collaborative_filtering
```

Among a wide range of algorithms to make recommendations, a popular one is matrix factorization or latent-feature methods

* Let $A$ be the user-item matrix where $A_{ij}=1$ if user $i$ and item $j$ is conncted. It can also be weighted to represent ratings.
* Each node has a position in a latent space of features. 
* A user $i$ with position $P_i$ rates an item $j$ with pisition $Q_j$ according to $P_i\cdot Q_j$.
* Infer nodes positions from the observed network. In matrix form, it is to find an approximate factorization of $A$.
$$ A\approx PQ $$
* Find $P$ and $Q$ that minimize the squared error.
$$ \sum_{(i,j)\in E} (A_{ij}-\sum_{k=1}^K P_{ik}Q_{kj})^2+\beta(||P||^2+||Q||^2) $$

```python
# This function and the following example is adapted from http://www.quuxlabs.com/blog/2010/09/matrix-factorization-a-simple-tutorial-and-implementation-in-python/#source-code
def matrix_factorization(R, P, Q, K, steps=5000, alpha=0.0002, beta=0.02):
    for step in xrange(steps):
        for i in xrange(len(R)):
            for j in xrange(len(R[i])):
                if R[i][j] > 0:
                    eij = R[i][j] - np.dot(P[i,:],Q[:,j])
                    for k in xrange(K):
                        P[i][k] = P[i][k] + alpha * (2 * eij * Q[k][j] - beta * P[i][k])
                        Q[k][j] = Q[k][j] + alpha * (2 * eij * P[i][k] - beta * Q[k][j])
        eR = np.dot(P,Q)
        e = 0
        for i in xrange(len(R)):
            for j in xrange(len(R[i])):
                if R[i][j] > 0:
                    e = e + pow(R[i][j] - np.dot(P[i,:],Q[:,j]), 2)
                    for k in xrange(K):
                        e = e + (beta/2) * (pow(P[i][k],2) + pow(Q[k][j],2))
        if e < 0.001:
            break
    return P, Q
```

```python
A = [[5,3,0,1],
     [4,0,0,1],
     [1,1,0,5],
     [1,0,0,4],
     [0,1,5,4]]

A = np.array(A)
N = len(A)
M = len(A[0])
K = 2
pd.DataFrame(A,columns=['Boyhood','The Grand Budapest Hotel','Whiplash','The Theory of Everything'],index=['Aaron','Bill','Eamon','Misha','Valentin'])
```

```python
P = np.random.rand(N,K)
Q = np.random.rand(K,M)

Ph, Qh = matrix_factorization(A, P, Q, K)
Ah = np.dot(Ph, Qh)
```
